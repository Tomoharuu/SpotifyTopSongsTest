```python
import pandas as pd
import plotly.express as px
df = pd.read_csv('https://raw.githubusercontent.com/Tomoharuu/SpotifyTopSongsTest/main/spotify_top_charts_22.csv')
display(df)
```

```python
df2 = df.copy()
df2['Artista Solo'] = df2['artist_names'].str.split(', ')
df2 = df2.explode('Artista Solo')
df2
```

```python
contagem1 = df2.copy()
contagem1 = contagem1[['Artista Solo', 'track_name', 'peak_rank','weeks_on_chart']]
contagem1
```

```python
valor_contagem1 = contagem1.value_counts('Artista Solo').reset_index()
valor_contagem1 = valor_contagem1.rename({'Artista Solo': 'Artista Solo', 0: 'Contagem'}, axis='columns').head(20).sort_values(by='Contagem')
valor_contagem1
```

```python
grafico1 = px.histogram(valor_contagem1, x='Contagem', y='Artista Solo')
grafico1.update_layout(title_text = 'Artistas mais frequentes nos rankings do Spotify', xaxis_title_text = 'Número de músicas', yaxis_title_text = 'Artista')
grafico1.show("png")
```

```python
contagem2 = contagem1.groupby(['Artista Solo']).mean().sort_values(by='weeks_on_chart', ascending=False).head(20).reset_index()
contagem2 = contagem2.drop(columns=['peak_rank']).sort_values(by='weeks_on_chart', ascending=True)
contagem2
```

```python
grafico2 = px.histogram(contagem2, x='weeks_on_chart', y='Artista Solo')
grafico2.update_layout(title_text='Média de semanas nos rankings do Spotify', xaxis_title_text='Média de semanas', yaxis_title_text='Artista')
grafico2.show("png")
```

```python
grafico3 = px.scatter(df2, x='weeks_on_chart', y='tempo', color='tempo', trendline='ols')
grafico3.update_layout(title_text='Popularidade de acordo com ritmo da música', xaxis_title_text='Semanas no Ranking', yaxis_title_text='Tempo')
grafico3.show("png")
```

```python
df2['Duração (min)'] = df2['duration_ms']
df2['Duração (min)'] = df2['Duração (min)'].div(60000).round(1)
df2
```

```python
grafico4 = px.scatter(df2, x='weeks_on_chart', y='Duração (min)', color='Duração (min)', trendline='ols')
grafico4.update_layout(title_text='Popularidade de acordo com a duração da música', xaxis_title_text='Semanas no Ranking', yaxis_title_text='Duração (min)')
grafico4.show("png")
```
